Loop Control

````python
# Pass

for i in range(6):
    print(i)
    if i < 2:
        print("eiei ^_^")
    else:
        pass
        print("oop?")
    print("go go go!")
````

````python
# Continue

for i in range(6):
    print(i)
    if i < 2:
        print("eiei ^_^")
    else:
        continue
        print("oop?")
    print("go go go!")
````

````python
# Break

for i in range(6):
    print(i)
    if i < 2:
        print("eiei ^_^")
    else:
        break
        print("oop?")
    print("go go go!")
````

2) Write a program that uses a `while` loop to find the smallest integer `n` such that <code>n<sup>2</sup></code> is greater than 12000.

````python
n = 1  # Start from 1

# Loop until n squared is greater than 12000
while n ** 2 <= 12000:
    n += 1

# Print the result
print("The smallest integer n such that n^2 > 12000 is:", n)
````

3) Write a program that gets a name forever or until `done` is entered.
- Create a variable `name` and assign it to an empty string.
- Use `while True:`
- Ask the user to enter a name.
- Keep asking until a non-blank/non-space alphabetical name is entered.
- Break the loop and print a greeting using `name`.

````python
# Create a variable name and assign it to an empty string
name = ""

# Loop forever until a valid name is entered
while True:
    name = input("Enter your name (or type 'done' to quit): ").strip()

    # Exit if user types 'done'
    if name.lower() == "done":
        print("Goodbye!")
        break

    # Check for non-empty, alphabetic name
    if name.isalpha():
        print("Hello,", name + "!")
        break
    else:
        print("Please enter a valid alphabetical name.")
````

8) Write a program that reads 10 integer values from the user and prints the highest value entered.

````python
# Initialize the highest value with a very small number
highest = None

# Read 10 integers from the user
for i in range(10):
    num = int(input(f"Enter integer #{i+1}: "))

    # Update the highest valu987e
    if highest is None or num > highest:
        highest = num

# Print the highest value entered
print("The highest value entered is:", highest)
````

9) Write a program that does the following:  
- Get a user input for the variable `first_name`.
- Create an empty string and assign it to the variable `new_name`.
- Iterate through all letters in `first_name`:
  - Add each letter in `new_name`.
  - Capitalize if letter is an `i` or `o`.
- Print the `new_name`.

````python
# Get input from user
first_name = input("Enter your first name: ")

# Create an empty string for the new name
new_name = ""

# Loop through each letter in first_name
for letter in first_name:
    if letter == 'i' or letter == 'o':
        new_name += letter.upper()  # Capitalize 'i' or 'o'
    else:
        new_name += letter  # Add the letter as is

# Print the modified name
print("Modified name:", new_name)
````
