# Time Series Forecasting

```python
from prophet import Prophet
```

```python
df_groupby = df[df['tpe'] == 'LAN_Retail'].groupby('date').agg(total_unit_sold = ('unit_sold', 'sum')).reset_index()
```

```python
df_groupby.head()
```

```python
df_groupby = df_groupby.rename(columns={'date': 'ds', 'total_unit_sold': 'y'})
```

```python
m = Prophet()
m.fit(df_groupby)
```

```python
future = m.make_future_dataframe(periods=10)
future.tail()
```

```python
forecast = m.predict(future)
forecast.head()
```

```python
fig1 = m.plot(forecast)
```

```python
fig2 = m.plot_components(forecast)
```
